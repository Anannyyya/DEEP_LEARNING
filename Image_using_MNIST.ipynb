```python
#example program on images using MNIST dataset
import matplotlib.pyplot as plt
from tensorflow import keras

#load MNIST dataset (predefined)
(x_train, y_train),(x_test,y_test)=keras.datasets.mnist.load_data()

#display functions for images
def show_images(images,labels, num=10):
    plt.figure(figsize=(10,5))
    for i in range(num):
      plt.subplot(2, 5, i+1) #print 10 images in 2 different lines
      plt.imshow(images[i], cmap='magma')
      plt.title(f"label :{labels[i]}")
      plt.axis('off')
      plt.show()
show_images(x_train, y_train)
```
